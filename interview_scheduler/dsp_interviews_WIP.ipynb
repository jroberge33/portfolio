```python
def read_data(filename, header = 0):
    ''' takes in a csv file, with an optional header command to specify which 
    row is the header and returns a dataframe with the designated row a 
    header'''
    
    df = pd.read_csv(filename, header = header)
    return df
```

```python
def rename_columns(df, original_column_names, new_column_names):
    ''' takes in a dataframe, original column names and new names 
    corresponding to those original columns in order to rename them, 
    returning a dataframe with these columns renamed '''
    
    # loop through all the columns in the original columns
    for i in range(len(original_column_names)):
        
        #rename them to correspond to the same column in the renamed list
        df.rename(columns={original_column_names[i]:new_column_names[i]}, 
                  inplace = True)
    return df
```

```python
def split_availability_columns(df, columns, days, hours, space = True):
    if space == True:
        for day in days:
            for hour in hours:
                df[f"{day}_{hour}_pm"] = df[f"{day}_availability"].str.contains(f"{int(hour)}-{int(hour) + 1} PM").astype(int)
    else:
        for day in days:
            for hour in hours:
                df[f"{day}_{hour}_pm"] = df[f"{day}_availability"].str.contains(f"{int(hour)}â€“{int(hour) + 1}PM").astype(int)
    return df
```

```python
def split_df_by_day(df, day, hours, pledge_classes):
    columns = ['interviewer_name', 'interviewer_pc']
    for hour in hours:
        columns.append(f"{day}_{hour}_pm")
    day_df = df[columns]
    day_df['interviewer_pc'] = pd.Categorical(day_df['interviewer_pc'], categories = pledge_classes, ordered = True)
    day_df.sort_values(by = 'interviewer_pc', inplace = True)
    return day_df
```

```python
def distribute_interviewers(day_df, num_interviews, day, hours, minutes, rooms):
    interviewer_dct = {}
    minute_tracking = {}
    pc_tracking = {}
    for hour in hours:
        interviewer_dct[f"{day}_{hour}_pm"] = {}
        for minute in minutes:
            minute_tracking[f"{day}_{hour}:{minute}_pm"] = []
            interviewer_dct[f"{day}_{hour}_pm"][minute] = {}
            for room in rooms:
                pc_tracking[f"{day}_{hour}:{minute}_pm_{room}"] = []
                interviewer_dct[f"{day}_{hour}_pm"][minute][room] = []
    interviewers_aggregate = []
    unfilled_interviews = 0
    for hour in hours:
        for minute in minutes:
            interviewers_ordered = list(day_df.loc[day_df[f"{day}_{hour}_pm"] == 1]['interviewer_name'])
            interviewers_pc = list(day_df.loc[day_df[f"{day}_{hour}_pm"] == 1]['interviewer_pc'])
            for j in range(3):
                for room in rooms:
                    i = 0
                    while len(list(interviewer_dct[f"{day}_{hour}_pm"][minute][room])) < 3 and len(interviewers_aggregate) < (num_interviews * 3):
                        try:
                            # print(f"TRY: {day}_{hour}:{minute}_room_{room}: {interviewers_ordered[i]}")
                            # print(interviewers_ordered)
                            # print(pc_tracking[f"{day}_{hour}:{minute}_pm_{room}"].count(interviewers_pc[i]))
                            # print(interviewers_aggregate.count(interviewers_ordered[i]))
                            if (pc_tracking[f"{day}_{hour}:{minute}_pm_{room}"].count(interviewers_pc[i]) <= 1) and (interviewers_aggregate.count(interviewers_ordered[i]) < 6):
                                interviewer_dct[f"{day}_{hour}_pm"][minute][room].append(interviewers_ordered[i])
                                minute_tracking[f"{day}_{hour}:{minute}_pm"].append(interviewers_ordered[i])
                                pc_tracking[f"{day}_{hour}:{minute}_pm_{room}"].append(interviewers_pc[i])
                                interviewers_aggregate.append(interviewers_ordered[i])
                                #print(f"SELECT: {day}_{hour}:{minute}_room_{room}: {interviewers_ordered[i]}")
                                interviewers_ordered.pop(i)
                                interviewers_pc.pop(i)
                                #print()
                            i += 1
                        except IndexError:
                            break

    return interviewer_dct
```

```python
def daily_dct_function(df, fit_interviews, case_interviews, fit_days, days, hours, minutes, rooms, pledge_classes):
    master_days_lst = []
    for day in days:
        day_df = split_df_by_day(df, day, hours, pledge_classes)
        if day in fit_days:
            day_dct = distribute_interviewers(day_df, round((int(fit_interviews) / 2), 0) + 1, day, hours, minutes, rooms)
            master_days_lst.append(day_dct)
        else:
            day_dct = distribute_interviewers(day_df, round((int(case_interviews) / 2), 0) + 1, day, hours, minutes, rooms)
            master_days_lst.append(day_dct)
    return master_days_lst
```

```python
def create_day_output_df(master_lst, day, days, hours, minutes, rooms, num_interviewers):
    day_index = int(days.index(day))
    day_interviews = master_lst[day_index]
    room_times = []
    interviewer_1 = []
    interviewer_2 = []
    interviewer_3 = []
    for hour in hours:
        for minute in minutes:
            for room in rooms:
                if day_interviews[f"{day}_{hour}_pm"][minute][room]:
                    room_times.append(f"{hour}:{minute}_pm_room_{room}")
    for key, value in day_interviews.items():
        hour = value
        for k, v in hour.items():
            minute = v
            for ke, val in minute.items():
                if val:
                    if len(list(val)) > 0:
                        interviewer_1.append(list(val)[0])
                    else:
                        interviewer_1.append(" ")
                    if len(list(val)) > 1:
                        interviewer_2.append(list(val)[1])
                    else:
                        interviewer_2.append(" ")
                    if len(list(val)) > 2:
                        interviewer_3.append(list(val)[2])
                    else:
                        interviewer_3.append(" ")
    dct = {"interviewer_1": interviewer_1, "interviewer_2": interviewer_2, "interviewer_3": interviewer_3}
    day_df = pd.DataFrame(data = dct, index = room_times)
    return day_df
```

```python
def split_availability_columns_apps(df, columns, days, hours):
    for day in days:
        for hour in hours:
            for minute in minutes:
                if minute == '40':
                    minute_string = f"{hour}:{minute} - {int(hour) + 1}:{int(minute) - 40} pm"
                else:
                    minute_string = f"{hour}:{minute} - {hour}:{int(minute) + 20} pm"
                df[f"{day}_{hour}:{minute}_pm"] = df[f"{day}_availability"].str.contains(minute_string).astype(int)
    return df
```

```python
def determine_date(df, fit_days, num_fits):
    applicants = list(df['full_name'])
    day_1_applicants = []
    either_applicants = []
    day_2_applicants = []
    for applicant in applicants:
        day_1_availability = list(df.loc[df['full_name'] == applicant][f'{fit_days[0]}_availability'])
        day_2_availability = list(df.loc[df['full_name'] == applicant][f'{fit_days[1]}_availability'])
        if len(day_1_availability[0].split(',')) > len(day_2_availability[0].split(',')):
            day_1_applicants.append(applicant)
        if len(day_1_availability[0].split(',')) == len(day_2_availability[0].split(',')):
            either_applicants.append(applicant)
        if len(day_1_availability[0].split(',')) < len(day_2_availability[0].split(',')):
            day_2_applicants.append(applicant)
    if num_fits % 2 == 0:
        day_1_spots = (num_fits // 2) - len(day_1_applicants)
    else:
        day_1_spots = (num_fits // 2) - len(day_1_applicants) + 1
    day_2_spots = (num_fits // 2) - len(day_2_applicants)
    for i in range(len(either_applicants)):
        if i < day_1_spots:
            day_1_applicants.append(either_applicants[i])
        else:
            day_2_applicants.append(either_applicants[i])
    for applicant in applicants:
        if applicant in day_1_applicants:
            df.loc[df['full_name'] == applicant, 'day_pattern'] = 'mon_wed'
        if applicant in day_2_applicants:
            df.loc[df['full_name'] == applicant, 'day_pattern'] = 'tue_thu'
    return df
```

```python
def fit_scheduling(df, fit_days, day_combos, hours, minutes, brothers_master_lst):
    scheduled_dct = {}
    for day in fit_days:
        day_apps = df.loc[df['day_pattern'] == f"{day}_{day_combos[day]}"]
        applicants = list(day_apps['full_name'])
        hours_dct = {}
        ordered_hours = []
        unsorted_applicants = []
        for hour in hours:
            available_apps = list(day_apps.loc[df[f'{day}_{hour}_pm']==1]['full_name'])
            hours_dct[hour] = len(available_apps)
        sorted_hours = sorted(hours_dct.items(), key=lambda x:x[1])
        for item in sorted_hours:
            ordered_hours.append(item[0])
        for hour in hours:
            for minute in minutes:
                scheduled_dct[f"{day}_{hour}:{minute}_pm"] = []
                
        applicants_availability = {}
        for applicant in applicants:
            applicant_availability_count = 0
            for hour in hours:
                if int(df.loc[df['full_name'] == applicant][(f"{day}_{hour}_pm")]) == 1:
                    applicant_availability_count += 1
            applicants_availability[applicant] = applicant_availability_count
        ordered_applicants = []
        sorted_applicants = sorted(applicants_availability.items(), key=lambda x:x[1])
        for item in sorted_applicants:
            ordered_applicants.append(item[0])
        
        finished_applicants = []
        for applicant in ordered_applicants:
            applicant_availability = []
            for hour in hours:
                if int(df.loc[df['full_name'] == applicant][(f"{day}_{hour}_pm")]) == 1:
                    applicant_availability.append(f"{day}_{hour}_pm")
                    
            for hour in hours:
                for minute in minutes:
                    if (f"{day}_{hour}_pm" in applicant_availability) and (len(scheduled_dct[f"{day}_{hour}:{minute}_pm"]) < 4) and (len(list(brothers_master_lst[int(fit_days.index(day))][f"{day}_{hour}_pm"][minute][int(len(list(scheduled_dct[f"{day}_{hour}:{minute}_pm"]))) + 1])) == 3):
                        scheduled_dct[f"{day}_{hour}:{minute}_pm"].append(applicant)
                        finished_applicants.append(applicant)
                        break
                    else:
                        unsorted_applicants.append(applicant)
                if applicant in finished_applicants:
                    break 
                    
        for applicant in unsorted_applicants:
            for hour in hours:
                for minute in minutes:
                    
                    if (len(scheduled_dct[f"{day}_{hour}:{minute}_pm"]) < 4) and (len(list(brothers_master_lst[int(fit_days.index(day))][f"{day}_{hour}_pm"][minute][int(len(list(scheduled_dct[f"{day}_{hour}:{minute}_pm"]))) + 1])) == 3):
                        scheduled_dct[f"{day}_{hour}:{minute}_pm"].append(applicant)
                        break 
                if applicant in finished_applicants:
                    break 
    
    return scheduled_dct
```

```python
def case_scheduling(df, case_days, day_combos, hours, minutes, brothers_master_lst, fit_schedule):
    scheduled_dct = {}
    for day in case_days:
        day_apps = df.loc[df['day_pattern'] == f"{day}_{day_combos[day]}"]
        applicants = list(day_apps['full_name'])
        
        hours_dct = {}
        ordered_hours = []
        unsorted_applicants = []
        for hour in hours:
            available_apps = list(day_apps.loc[df[f'{day}_{hour}_pm']==1]['full_name'])
            hours_dct[hour] = len(available_apps)
        sorted_hours = sorted(hours_dct.items(), key=lambda x:x[1])
        for item in sorted_hours:
            ordered_hours.append(item[0])
        for hour in hours:
            for minute in minutes:
                scheduled_dct[f"{day}_{hour}:{minute}_pm"] = []
                
        applicants_availability = {}
        for applicant in applicants:
            applicant_availability_count = 0
            for hour in hours:
                if int(df.loc[df['full_name'] == applicant][(f"{day}_{hour}_pm")]) == 1:
                    applicant_availability_count += 1
            applicants_availability[applicant] = applicant_availability_count
        ordered_applicants = []
        sorted_applicants = sorted(applicants_availability.items(), key=lambda x:x[1])
        for item in sorted_applicants:
            ordered_applicants.append(item[0])
        
        finished_applicants = []
        for applicant in ordered_applicants:
            applicant_availability = []
            for hour in hours:
                if int(df.loc[df['full_name'] == applicant][(f"{day}_{hour}_pm")]) == 1:
                    applicant_availability.append(f"{day}_{hour}_pm")
                    
            for hour in hours:
                for minute in minutes:
                    if (f"{day}_{hour}_pm" in applicant_availability) and (len(scheduled_dct[f"{day}_{hour}:{minute}_pm"]) < 4) and (len(list(brothers_master_lst[int(fit_days.index(day))][f"{day}_{hour}_pm"][minute][int(len(list(scheduled_dct[f"{day}_{hour}:{minute}_pm"]))) + 1])) == 3):
                        scheduled_dct[f"{day}_{hour}:{minute}_pm"].append(applicant)
                        finished_applicants.append(applicant)
                        break
                    else:
                        unsorted_applicants.append(applicant)
                if applicant in finished_applicants:
                    break 
                    
        for applicant in unsorted_applicants:
            for hour in hours:
                for minute in minutes:
                    if (len(scheduled_dct[f"{day}_{hour}:{minute}_pm"]) < 4) and (len(list(brothers_master_lst[int(fit_days.index(day))][f"{day}_{hour}_pm"][minute][int(len(list(scheduled_dct[f"{day}_{hour}:{minute}_pm"]))) + 1])) == 3):
                        scheduled_dct[f"{day}_{hour}:{minute}_pm"].append(applicant)
                        break 
                if applicant in finished_applicants:
                    break 
    
    return scheduled_dct
```
